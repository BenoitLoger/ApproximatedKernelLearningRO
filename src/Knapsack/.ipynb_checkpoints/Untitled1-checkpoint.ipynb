```julia
function update_SV(X::Array{Float64,2}, Q::Array{Float64,2},alpha_ini::Array{Float64,1},beta_sel::Array{Float64,1}, selected::Array{Int,1}, SV::Array{Int,1}, used,dist_set::Array{Int,1},max_D;time_limit=1800)
	println("length used = ",length(used))
    m = Model(CPLEX.Optimizer)
	@variable(m, D[i in dist_set] >= 0) # Maximum distance
    @variable(m, 0 <= beta[i in used] <= 1)
    @variable(m, y[i in used], Bin)

    for i in selected
        set_start_value(y[i], 1)
		set_start_value(beta[i], beta_sel[i])
    end
	for i in setdiff(used,selected)
		set_start_value(y[i],0)
		set_start_value(beta[i], 0.0)
	end

	for i in dist_set
		set_start_value(D[i], max_D[i])
	end

    @objective(m, Max, length(selected)-1 - sum(y[i] for i in used))

    @constraint(m, [i in dist_set], D[i] >= sum((alpha_ini[ii])*norm(Q*(X[:,i]-X[:,ii]),1) for ii in SV)
        - sum((beta[ii])*norm(Q*(X[:,i]-X[:,ii]),1) for ii in used))
    @constraint(m, [i in dist_set], D[i] >= sum((beta[ii])*norm(Q*(X[:,i]-X[:,ii]),1) for ii in used)
        - sum((alpha_ini[ii])*norm(Q*(X[:,i]-X[:,ii]),1) for ii in SV))
	# @constraint(m,[g in 1:length(groups)], sum(y[gg] for gg in [k for k in groups[g] if k in used]) <= 1)
	# @constraint(m, [i in dist_set], D[i] <= maximum(max_D))
	@constraint(m, sum(D[i] for i in dist_set) <= sum(max_D))
	@constraint(m, [i in used], y[i] >= beta[i])
	# @constraint(m, [i in setdiff(used,selected)], y[i] <= 0)
    @constraint(m,sum(y[i] for i in used) <= length(selected))

	# Options for solving
	set_optimizer_attribute(m, "CPXPARAM_Threads", 4)
	set_time_limit_sec(m,time_limit)
	# set_silent(m)
	solving_time = (@timed optimize!(m))[2] # Solving

	selected::Array{Int,1} = [i for i in used if value.(m[:beta])[i] > 0]
	weights = zeros(Float64,length(X[1,:]))
	for i in selected
		weights[i] = value.(m[:beta])[i]
	end
	return selected, weights, value.(m[:D]), solving_time
end
function phase_2(svc::SVC, approx::SVC, groups::Array{Array{Int,1},1}, max_distance;time_limit=60)
	N::Int = size(svc.Data,2) # Get the number of sample points in the data set
	used = svc.SV
# 	if length(svc.SV) > 200
# 		sel, groups = subset_selection(svc.Data,svc.Q,svc.SV,svc.BSV,200)
# 		used = union(sel,approx.SV)
# 	end
	selected, beta, max_distance, LP_time = update_SV(svc.Data,svc.Q,svc.Alpha,approx.Alpha,approx.SV,svc.SV,used,collect(1:N),max_distance,time_limit=time_limit)
	selected, beta, max_distance, LP_time = Min_Sum(svc.Data,svc.Q,svc.Alpha,selected,svc.SV,collect(1:N))
	ret = SVC(svc.Data,svc.nu)
	ret.SV = selected
	ret.Alpha = beta
	ret.BSV = svc.BSV
	ret.Theta = svc.Theta
	return ret, max_distance
end
phase2, max_distance = phase_2(svc,approx,[[1]],dist,time_limit=60)
```
