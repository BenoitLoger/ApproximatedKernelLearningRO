```julia
function ASVC_quality(m::Int,name::String,r::Float64)
	final_df = DataFrame(m=Int[],nu=Float64[],index=Int[],K=Int[],sub_SV=Int[],neg=Float64[],pos=Float64[])
	for nu in [0.15,0.25,0.35,0.45,0.55,0.60,0.70,0.75,0.80,0.85,0.90,0.95,0.99]
        for r in 1:0.5:2.5
			K = Int(m*r)
			loc_df = DataFrame(CSV.File(string(INSTANCE_PATH,name,"/ASVC_quality_",nu,"_",K,".csv")))
			for row in eachrow(loc_df)
				push!(final_df,row)
			end
		end
	end
	return final_df
end

plt = plot(ylim=(0,3.5),xlabel="Value of the ratio K/m",ylabel="Percentage of Type II error")
for m in 10:10:50
    df = ASVC_quality(m,string("gamma_",m,"_1"))
#     k_m = [r for r in 1:0.5:2.5]
    plt = plot!([row.K/m for row in eachrow(df) if row.m == m && row.nu == 0.35],[row.neg for row in eachrow(df) if row.m == m && row.nu == 0.35],label = string("m = ",m),markershape=:circle)
end
display(plt)
savefig(plt,"Type_2.png")
```

```julia
saved = (10,0.15)
test = false
for row in eachrow(df)
    if saved != (row.m,row.nu) 
        saved = (row.m,row.nu) 
        test = false
    end
    loc_df = DataFrame(CSV.File(string(INSTANCE_PATH,"gamma_",row["m"],"/ASVC*_quality_",row.nu,"_",row.K,".csv")))
    min_diff = 100
    min_ind = 0
    for loc_row in eachrow(loc_df)
        loc_error = loc_row.neg+loc_row.pos
        if abs(loc_error-row.error) < min_diff
            min_diff = abs(loc_error-row.error)
            min_ind = loc_row.index
        end
    end
    if row.error <= 1.0 && !test
        println("./../../../julia-1.7.2/bin/julia main.jl ",row.m," ",row.nu," ",row.K," ",min_ind," > out/phase2_",row.m," ",row.nu," ",row.K," ",min_ind," &")
        test = true
    end
end
```

```julia
nb_eq = zeros(Int,7)
tot_optimal = 0
i = 1
for r in 1:0.5:4
    for m in [10,20,30,40]
        name = string("gamma_",m,"_1")
        svc_df = load_sol(INSTANCE_PATH,string(name,"/SVC.txt"))
        for svc_row in eachrow(svc_df)
            if svc_row.status == "Optimal"
                tot_optimal += 1
                K = r*m
                asvc_df = load_sol(INSTANCE_PATH,string(name,"/ASVC_",Int(K),".txt"))
                for row in eachrow(asvc_df)
                    if row.nu == svc_row.nu
                        if row.solution == svc_row.solution
                            nb_eq[i] += 1
                        end
                    end
                end
            end
        end
    end
    i += 1
end
```

```julia
name = "gamma_30_1"
K = 60
# Loading Instance and creating test set
inst = Instance(name,path=INSTANCE_PATH)
# println([quantile(inst.w[i,:],0.95) for i in 1:20])
X_test = inst.dist.generate(10000)
m = length(inst.c)
println([mean(X_test[i,:]) for i in 1:m])

# Deterministic model
model = Deterministic(inst,variable_type="Binary")
det, t = solve_model(model)
obj_det, res_det, load= evaluate(inst,det,X_test)
# Budget based model
# df_bert = load_bert(path,string(name,"/Bertsimas.txt"))
gammas = collect(0.0:0.05:5)
obj_B = Array{Float64,1}(undef,length(gammas))
sat_B = Array{Float64,1}(undef,length(gammas))
pro_B = Array{Float64,1}(undef,length(gammas))
# # plt = plot()
# times = Array{Float64,1}(undef,length(gammas))
for i in 1:length(gammas)
    B = Bertsimas(inst,gammas[i],variable_type="Binary")
    B_sol, s, s, s = solve_model(B)
    obj_B[i], sat_B[i],load = evaluate(inst,B_sol,X_test)
    pro_B[i] = mean(load)
    plt = plot(obj_B,sat_B.*100, markershape=:circle,label="Budget",legend = :bottomleft)
end
df_svc30 = load_sol(INSTANCE_PATH,string(name,"/SVC.txt"))
println(df_svc30)
nus = df_svc30[!,"nu"]
obj_SVC = Array{Float64,1}(undef,length(nus))
sat_SVC = Array{Float64,1}(undef,length(nus))
pro_SVC = Array{Float64,1}(undef,length(nus))
for i in 1:length(nus)
    svc_sol =  parse.(Float64,split(df_svc30[i,"solution"][2:end-1],","))
    obj_SVC[i], sat_SVC[i],load = evaluate(inst,svc_sol,X_test)
    pro_SVC[i] = mean(load)
end
plt = plot(obj_B,sat_B.*100, markershape=:circle,label="Budget",legend = :bottomleft)
opt = df_svc30[!,"status"]
plt = plot!(obj_SVC,sat_SVC.*100,label=nothing)
opt_legend = false
nopt_legend=false
for i in 1:length(nus)
    if opt[i] == "Optimal"
        if opt_legend
            plt = scatter!([obj_SVC[i]],[sat_SVC[i]*100],label=nothing,color=:red,markershape=:rect)
        else
            plt = scatter!([obj_SVC[i]],[sat_SVC[i]*100],label="SVC gap = 0",color=:red,markershape=:rect)
            opt_legend = true
        end
    else
        if nopt_legend
            plt = scatter!([obj_SVC[i]],[sat_SVC[i]*100],label=nothing,color=:red,markershape=:dtri)
        else
            plt = scatter!([obj_SVC[i]],[sat_SVC[i]*100],label="SVC gap > 0",color=:red,markershape=:dtri)
            nopt_legend = true
        end
    end
    println(sat_SVC)
end
display(plt)

# for i in 1:10
# 	df_svc30 = load_sol(INSTANCE_PATH,string(name,"/ASVC*_",K,"_",i,".txt"))
# 	# println(df_svc)
# 	nus = df_svc30[!,"nu"]
# 	obj_SVC = Array{Float64,1}(undef,length(nus))
# 	sat_SVC = Array{Float64,1}(undef,length(nus))
# 	pro_SVC = Array{Float64,1}(undef,length(nus))
#
# 	for i in 1:length(nus)
# 	    svc_sol =  parse.(Float64,split(df_svc30[i,"solution"][2:end-1],","))
# 	    obj_SVC[i], sat_SVC[i] = evaluate(inst,svc_sol,X_test)
# 	end
# 	plt = plot!(obj_SVC,sat_SVC.*100,label="ASVC*",markershape=:utri)
# end

df_svc30 = load_sol(INSTANCE_PATH,string(name,"/ASVC_",K,".txt"))
sort!(df_svc30,["nu"])
println(df_svc30)
nus = df_svc30[!,"nu"]
obj_SVC = Array{Float64,1}(undef,length(nus))
sat_SVC = Array{Float64,1}(undef,length(nus))
pro_SVC = Array{Float64,1}(undef,length(nus))

for i in 1:length(nus)
    svc_sol =  parse.(Float64,split(df_svc30[i,"solution"][2:end-1],","))
    obj_SVC[i], sat_SVC[i],load = evaluate(inst,svc_sol,X_test)
    pro_SVC[i] = mean(load)
end
x_min = minimum(obj_SVC)-0.05*minimum(obj_SVC)
x_max = maximum(obj_SVC)+0.05*maximum(obj_SVC)
y_min = minimum(sat_SVC)*100-0.05*minimum(sat_SVC)*100
plt = plot!(obj_SVC,sat_SVC.*100,label="ASVC",markershape=:utri,xlabel="Value of the objective function",ylabel="Percentage of constraint satisfaction",xlim=(x_min,x_max),ylim=(y_min,102),color=:green)
display(plt)
savefig(plt,"KP_30.png")
```
