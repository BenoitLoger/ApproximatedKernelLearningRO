```julia
cors = []
cs = []
m = 10
dist_weight = Uniform(50,100)
dist_cost = Uniform(17,77)
act = 0
for c in collect(0.01:0.1:0.99)
    input = rand(MvNormal([0,0],
        [1.0 c
         c 1.0]),m)
    cdf_w = cdf(Normal(),input[1,:])
        cdf_c = cdf(Normal(),input[2,:])
        weights = quantile.(dist_weight,cdf_w)
        costs = quantile.(dist_cost,cdf_c)
        act = cor(weights,costs)

    while abs(act-c) >0.01
        input = rand(MvNormal([0,0],
        [1.0 c
         c 1.0]),m)
        cdf_w = cdf(Normal(),input[1,:])
        cdf_c = cdf(Normal(),input[2,:])
        weights = quantile.(dist_weight,cdf_w)
        costs = quantile.(dist_cost,cdf_c)
        act = cor(weights,costs)
    end
    
    

    push!(cs,c)
    push!(cors,cor(weights,costs))
end
```

```julia
plot(cs,cors,ylim=(0,1))
```

```julia
function wc(d1,d2,Cor::Float64;epsilon=0.01)
    act = 0
    input = rand(MvNormal([0,0],
        [1.0 Cor
         Cor 1.0]),m)
        cdf_w = cdf(Normal(),input[1,:])
        cdf_c = cdf(Normal(),input[2,:])
        weights = quantile.(d1,cdf_w)
        costs = quantile.(d2,cdf_c)
        act = cor(weights,costs)
        
        while abs(act-Cor) >epsilon
            input = rand(MvNormal([0,0],
            [1.0 Cor
             Cor 1.0]),m)
            cdf_w = cdf(Normal(),input[1,:])
            cdf_c = cdf(Normal(),input[2,:])
            weights = quantile.(d1,cdf_w)
            costs = quantile.(d2,cdf_c)
            act = cor(weights,costs)
        end
    return weights, costs, act
end
cors = []
cs = []
m = 30
dist_weight = Uniform(50,100)
dist_cost = Uniform(17,77)
for c in collect(0.0:0.1:0.9)
    w,cos,Cor = wc(dist_weight, dist_cost,c)
    plt = scatter(w,cos)
    savefig(plt,string(c,".png"))
    push!(cors,c)
    push!(cs,Cor)
end
```

```julia
plot(cs,cors,ylim=(0,1))
```
